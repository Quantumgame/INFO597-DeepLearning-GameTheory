```python
import pandas as pd
import numpy as np
def f(row):
    if row['Adj Close'] - row['Open'] < 0: return -1
    else: return 1
```

```python
df = pd.read_csv('data/GOOGLu.csv')
df['Utility'] = df.apply(f, axis=1)
df.head()
```

```python
data = df.drop('Utility', 1)
data = data.drop('Date',1).values
target = df['Utility'].values
```

### Playing with Stuff

```python
from random import shuffle
train_input = ['{0:020b}'.format(i) for i in range(2**20)]
shuffle(train_input)
train_input = [map(int,i) for i in train_input]

ti  = []
for i in train_input:
    temp_list = []
    for j in i:
        temp_list.append([j])
    ti.append(np.array(temp_list))
train_input = ti
```

```python
#print(train_input[0])
train_input[0].shape
```

```python
train_output = []
for i in train_input:
    count = 0
    for j in i:
        if j[0] == 1:
            count+=1
    temp_list = ([0]*21)
    temp_list[count]=1
    train_output.append(temp_list)
```
